```python
# PREPROCESSING

# You get some data and the data is NOT perfect to be used right away with you model
# So you wanna train...
# let's say a K neighbors classifier on your data, but the data is NOT perfect
# Some of the features mmoght not even be numerical. Some of the features might not be scaled yet.
# this is important for a distance-based classification algorithm

# So you need to pre-processing the data. you need to some stuff with the data before it can be used
# for training and evaluating the model

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# It basically scales the data to a specific format.
```

```python
X, y = load_iris(return_X_y=True)
# To make all features equally important, what we do is we scale them down to the same variance.
# So the unit variance essentially, and we can do it different ways.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# we gonna do a fit transform
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import numpy as np

(X_train - np.mean(X_train, axis=0)) / np.std(X_train, axis=0)
# otherwise we're taking the mean across all these values here
```
