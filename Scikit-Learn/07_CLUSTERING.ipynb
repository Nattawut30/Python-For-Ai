```python
# CLUSTERING (UNSUPERVISED LEARNING)

# You have a data and the data is unlabeled.
# You don't have any classes or something like that

# Now you wanna cluster the data into reasonable clusters.
# KMeans clustering 

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, _ = make_blobs(n_samples=5000, centers=5, random_state=10)

X

# You can use random_state in a bunch of different functions
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_scaled[:, 0], X_scaled[:, 1])

# obviously since we generated a blobs datasets
# These are the 5 diffrent clusters
# This can be much more noisy and confusing
```

```python
kmeans = KMeans(n_clusters=5)

kmeans.fit(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_)
# underscore because in scikit-learn indicates this is a trianed parameter.
# or something that is trained.
```

```python
# KMeans works by seperate data inti Shperical clusters
# Sometimes is not works with some data
# DBSSCAN or Spectral Clustering is better

from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan.labels_)
```
