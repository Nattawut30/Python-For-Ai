```python
# PCA (UNSEPERVISED LEARNING)
# Principle Component Analysis

# you take a high-dimensional datasets
# you have a thousand features and you wanna reduce dimensionality.
# So you wanna end up with 20 or 30 features
# You will go into this n-dimensional space and you try to find the directions in the space.

# These are not necessarily features. They can be new directions. A combinations of directions.
# you find it and explain the most variance and you reduct your datasets down to these directions.

# This can be used for example as a pre-processing step to reduct complexity or dimensionality before doing something else.

from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X, y = fetch_openml('mnist_784', return_X_y=True) # 28*28 = 784 pixels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Each pixels is an activation value and each pixel is a feature.
# 784 features to predict 1 out of 10 classes
```

```python
# Check the loss of information. the loss of performance
# how much it's affect your performance?

pca = PCA(n_components=10)

X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
X_train.shape
```

```python
X_train_reduced.shape
```

```python
clf = LogisticRegression(max_iter=100) 
# limiting the number of iterations in this case to not have this run too long.
clf.fit(X_train, y_train)

print(clf.score(X_test, y_test))
```

```python
# Let's try to reduce this time for faster.

clf = LogisticRegression(max_iter=100) 
clf.fit(X_train_reduced, y_train)

print(clf.score(X_test_reduced, y_test))

# It's still 80%+ still lookin good
```
